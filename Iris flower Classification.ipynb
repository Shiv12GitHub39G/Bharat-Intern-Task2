```python
x= df1[['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y= df1['Species']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# Training Using Logistic Regression Model....

from sklearn.linear_model import LogisticRegression
```

```python
l1 = LogisticRegression()
l1.fit(x,y)

# Train the model with the training set
l1.fit(x_train,y_train)
```

```python
predictions = l1.predict(x)
```

```python
Scores = pd.DataFrame({'Actual':y,'Predictions':predictions})
Scores.head()  # Predicting Different Species of Iris Flower
```

```python
Scores.tail()
```

```python
y_test_hat=l1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score #Checking the Accuracy of our Trained Model
print(accuracy_score(y_test,y_test_hat)*100,'%')
```
